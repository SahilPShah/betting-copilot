```python
import pandas as pd
import time

def get_pitching_logs(team, year):
    url = f"https://www.baseball-reference.com/teams/{team}/{year}-pitching-gamelogs.shtml"
    try:
        tables = pd.read_html(url)
        df = tables[0]
        df['team'] = team
        df['year'] = year
        df['log_type'] = 'pitching'
        print(f"✅ Pitching {team} {year} — {len(df)} rows, columns: {df.columns.tolist()}")
        print(df.head(3))
        return df
    except Exception as e:
        print(f"❌ Pitching {team} {year}: {e}")
        return None

def get_batting_logs(team, year):
    url = f"https://www.baseball-reference.com/teams/{team}/{year}-batting-gamelogs.shtml"
    try:
        tables = pd.read_html(url)
        df = tables[0]
        df['team'] = team
        df['year'] = year
        df['log_type'] = 'batting'
        print(f"✅ Batting {team} {year} — {len(df)} rows, columns: {df.columns.tolist()}")
        print(df.head(3))
        return df
    except Exception as e:
        print(f"❌ Batting {team} {year}: {e}")
        return None

# Test one team and year first
pitching = get_pitching_logs('ATL', 2024)
time.sleep(3)
batting = get_batting_logs('ATL', 2024)
```

```python
import pandas as pd

# Test different URL formats
urls = [
    "https://www.baseball-reference.com/teams/ATL/2024-batting-gamelogs.shtml",
    "https://www.baseball-reference.com/teams/ATL/2024-schedule-scores.shtml",
    "https://www.baseball-reference.com/teams/ATL/2024.shtml",
]

for url in urls:
    try:
        tables = pd.read_html(url)
        print(f"✅ {url}")
        print(f"   Found {len(tables)} tables")
        for i, t in enumerate(tables):
            print(f"   Table {i}: {t.shape} — cols: {t.columns.tolist()[:10]}")
    except Exception as e:
        print(f"❌ {url}: {e}")
    import time
    time.sleep(3)
```

```python
import pandas as pd
import time

def get_splits(team, year):
    url = f"https://www.baseball-reference.com/teams/{team}/{year}-splits.shtml"
    try:
        tables = pd.read_html(url)
        print(f"✅ {team} {year} — Found {len(tables)} tables")
        for i, t in enumerate(tables):
            print(f"\n  Table {i}: {t.shape}")
            print(f"  Columns: {t.columns.tolist()}")
            print(t.head(3))
        return tables
    except Exception as e:
        print(f"❌ {team} {year}: {e}")
        return None

# Test one team and year
tables = get_splits('ATL', 2024)
```

```python
import pandas as pd
from pybaseball import team_batting, team_pitching
import time

def get_team_batting(year):
    try:
        df = team_batting(year)
        print(f"✅ Batting {year} — {len(df)} rows")
        print(f"  Columns: {df.columns.tolist()}")
        print(df[['Team', 'OPS', 'AVG', 'OBP', 'SLG', 'wRC+', 'BB%', 'K%']].head(5))
        return df
    except Exception as e:
        print(f"❌ Batting {year}: {e}")
        return None

def get_team_pitching(year):
    try:
        df = team_pitching(year)
        print(f"✅ Pitching {year} — {len(df)} rows")
        print(f"  Columns: {df.columns.tolist()}")
        print(df[['Team', 'ERA', 'WHIP', 'K/9', 'BB/9', 'xFIP', 'FIP']].head(5))
        return df
    except Exception as e:
        print(f"❌ Pitching {year}: {e}")
        return None

# Test 2024
print("=== BATTING ===")
batting = get_team_batting(2024)
time.sleep(3)

print("\n=== PITCHING ===")
pitching = get_team_pitching(2024)
```

```python
import requests
import os
from dotenv import load_dotenv

load_dotenv("/Users/sahilshah/betting-copilot/.env")
API_KEY = os.getenv("ODDS_API_KEY")

# Test historical access with a known 2024 date
url = "https://api.the-odds-api.com/v4/historical/sports/baseball_mlb/odds"

params = {
    "apiKey": API_KEY,
    "regions": "us",
    "markets": "h2h,spreads",
    "oddsFormat": "american",
    "date": "2024-07-15T00:00:00Z",
    "bookmakers": "draftkings"
}

response = requests.get(url, params=params)
print("Status:", response.status_code)
print("Requests used:", response.headers.get('X-Requests-Used'))
print("Requests remaining:", response.headers.get('X-Requests-Remaining'))

data = response.json()
print("\nKeys:", list(data.keys()))
print("\nSample game:", str(data)[:500])
```

```python
import json

# Pretty print one game to see full odds structure
data = response.json()
if data['data']:
    print(json.dumps(data['data'][0], indent=2))
```

```python
# Pull a regular season date to see real team names
params['date'] = "2024-05-15T00:00:00Z"
response = requests.get(url, params=params)
data = response.json()

# Extract all unique team names
teams = set()
for game in data['data']:
    teams.add(game['home_team'])
    teams.add(game['away_team'])

for t in sorted(teams):
    print(t)
```
